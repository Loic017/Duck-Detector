# Process Images


```python
import os
import pandas as pd
import numpy as np
import PIL
from PIL import Image
```

```python
IMAGE_FOLDER = 'images/2_cropped_images'
IMAGE_FILE_TYPE = 'jpg'
```

```python
def get_image_files(IMAGE_FOLDER, IMAGE_FILE_TYPE):
    image_files = []
    for file in os.listdir(IMAGE_FOLDER):
        if file.endswith(IMAGE_FILE_TYPE):
            image_files.append(file)
    return image_files

def load_image_files(image_files, IMAGE_FOLDER):
    images = []
    for file in image_files:
        image = Image.open(IMAGE_FOLDER + '/' + file)
        images.append(image)
    return images
```

```python
image_files = get_image_files(IMAGE_FOLDER, IMAGE_FILE_TYPE)
images = np.array(load_image_files(image_files, IMAGE_FOLDER))
images.shape
```

Turn nparrays into images and resize 8 times smaller (208x504)


```python
sm_images = []
for image in images:
    im = Image.fromarray(image)
    sm_images.append(im.resize((im.width//8, im.height//8)))
```

Show one image


```python
image_number_x = 1
image_example = sm_images[image_number_x]
# show image[image_number_x]
display(image_example)
```

# Save New Images


```python
for i, image in enumerate(sm_images):
    image.save(f"images/3_resized_images/image_{i}.jpg")
```
